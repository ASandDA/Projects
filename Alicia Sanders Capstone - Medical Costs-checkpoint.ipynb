```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

```python
#exploring the data
df = pd.read_csv("Medical_Cost.csv")
df.head()
df.describe()
```

```python
#replacing y, n with 1, 0
df = df.replace({"smoker": {"yes":1,"no":0}})
df = df.replace({"sex": {"male":1,"female":0}})
df.head()
```

```python
#How many patients are in each region? What region has the highest concentration of patients?
region = df["region"].value_counts()
region
```

```python
region.plot(kind="bar",title="Total Patients per Region")
plt.xlabel("Region")
plt.ylabel("# of Patients")
plt.savefig("Cappic_Region")
plt.show()
```

```python
#What percent of total patients are in each region?
region_perc = df["region"].value_counts(normalize=True)
region_perc
```

```python
region_perc.plot(kind="pie",title="Regional share of patients")
plt.savefig("Cappic_Regionperc")
plt.show()
```

```python
#Looking at Medical Cost Distribution
charge_distr = df.charges.plot(kind="hist",bins=7,title="Distribution of Medical Costs")
plt.xlabel("Medical Costs")
plt.savefig("charge_distr")
plt.show()
#Over 75% of patients have medical costs that are $20,000 or less. Over half have medical costs less than $10,000
```

```python
#What are the total medical costs per region?
region_costs = df.groupby("region")["charges"].sum().sort_values(ascending=False)
region_costs.round()
```

```python
region_costs.plot(kind="bar",title="Total Medical Costs per Region in Millions",color=["red","orange","blue","green"])
plt.ylabel("Medical Costs (M)")
plt.savefig("Cappic_Regioncosts")
plt.show()
```

```python
#Does having children/multiple children affect medical costs?
kids = df.groupby("children")["charges"].mean().sort_values(ascending=False).round()
kids
```

```python
kids.plot(kind="bar",title="Average Medical Costs per # of Children",color=["red","orange","blue","yellow","green","purple"])
plt.xlabel("# of children")
plt.ylabel("Avg. Costs")
plt.savefig("kids")
plt.show()
```

```python
#Does smoking affect medical costs?
Smoke_status = df["smoker"].value_counts()
Smoke_status
```

```python
Smoke_costs = df.groupby("smoker")["charges"].mean().sort_values(ascending=False).round()
Smoke_costs
Smoke_costs.plot(kind="bar",title="Average Costs for Smoker vs. Non-Smoker",color=["red","green"])
plt.xlabel("Does Patient Smoke ?")
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_Smokecosts")
plt.show()
#We can see that patients who smoke, although only a small subset of the entire population, represent an outsized portion of average costs. 
```

```python
#creating a new column of Age groups
def Age_group(row):
    if row <= 18:
        result = "Child (0-18)"
    elif row <= 44:
        result = "Younger Adult Group (19-44)"
    else:
        result = "Older Adult Group (45-64)"
    return result
df["Age Group"] = df["age"].apply(Age_group)
df.head()
```

```python
#Which age group has the highest average medical costs?
Age_Costs = df.groupby("Age Group")["charges"].mean().sort_values(ascending=False).round()
Age_Costs
```

```python
Age_Costs.plot(kind="bar",title="Average Costs per Age Group",color=["red","blue","green"])
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_Agecosts")
plt.show()
```

```python
#Which gender has the highest average medical costs?
Gender_Costs = df.groupby("sex")["charges"].mean().sort_values(ascending=False).round()
Gender_Costs
```

```python
Gender_Costs.plot(kind="bar",title="Average Costs per Gender",color=["blue","pink"])
plt.xlabel('Gender')
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_Gendercosts")
plt.show()
```

```python
#Creating a classification for weight
def Weight_Group(row):
    if row < 18.5:
        result = "Underweight"
    elif row <= 24.9:
        result = "Healthy Weight"
    elif row < 30:
        result = "Overweight"
    else:
        result = "Obese"
    return result
df["Weight_Group"] = df["bmi"].apply(Weight_Group)
df.head()
```

```python
#Which weight classification has the highest average medical costs. 
Weight_Costs = df.groupby("Weight_Group")["charges"].mean().sort_values(ascending=False).round()
Weight_Costs
```

```python
Weight_Costs.plot(kind="bar",title="Average Costs per Weight Group",color=["red","orange","blue","green"])
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_Weightcosts")
plt.show()
```

```python
#creating a df to use for inferential analysis. 
df_predictors = df.loc[:,["age","sex","children","bmi","smoker","region","charges"]]
df_predictors
```

```python
#Inferential - Let's take a look at which variables seem to have the highest affect on medical costs.
df_predictors.reset_index()
correlation = df_predictors.corr()
heatmap = sns.heatmap(correlation,annot=True,cmap="Greens")
plt.savefig("Cappic_heatmap")
plt.show()
```

```python
#We can see that smoking has the highest correlation with medical costs. 
cost_correlation = correlation["charges"].sort_values().plot(kind="bar",title="Affect of Variables on Medical Cost")
plt.xlabel("predictors")
plt.ylabel("Correlation")
plt.savefig("Cappic_costcorr")
plt.show()
```

```python
#Regression Analysis
List_of_predictors = ["age","children","sex","smoker","bmi"]
independent_variables = df[List_of_predictors]
dependent_variable = df["charges"]
independent_variables = sm.add_constant(independent_variables)
regression = sm.OLS(dependent_variable,independent_variables).fit()
regression.summary()
```

```python
#R squared value of .75 which is a reasonable goodness of fit. 
regression.params
#It's clear that smoking is a strong indicator of higher medical costs, and increases costs by ~24,000. 
```

```python
#Now, looking at regions, which region has the highest medical costs of older adult male smokers?
Older_Male_Smoker = df[df["age"]>=45]
Older_Male_Smoker = Older_Male_Smoker[Older_Male_Smoker["sex"]==1]
Older_Male_Smoker = Older_Male_Smoker[Older_Male_Smoker["smoker"]==1]
OMS = Older_Male_Smoker.groupby("region")["charges"].mean().sort_values(ascending=False).round()
OMS
```

```python
OMS.plot(kind="bar",title="Avg. Medical Cost per Region of Older Adult Male Smokers",color=["red","orange","blue","green"])
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_OMS")
plt.show()
```

```python
#Which region has the highest medical costs of older adult female smokers?
Older_Female_Smoker = df[df["age"]>=45]
Older_Female_Smoker = Older_Female_Smoker[Older_Female_Smoker["sex"]==0]
Older_Female_Smoker = Older_Female_Smoker[Older_Female_Smoker["smoker"]==1]
OFS = Older_Female_Smoker.groupby("region")["charges"].mean().sort_values(ascending=False).round()
OFS
```

```python
OFS.plot(kind="bar",title="Avg. Medical Cost per Region of Older Adult Female Smokers",color=["red","orange","blue","green"])
plt.ylabel("Avg. Costs")
plt.savefig("Cappic_OFS")
plt.show()
```

```python
#Exploring any difference for young female smokers
Younger_Adult_Female_Smoker = df[df["age"]<45]
Younger_Adult_Female_Smoker = Younger_Adult_Female_Smoker[Younger_Adult_Female_Smoker["sex"]==0]
Younger_Adult_Female_Smoker = Younger_Adult_Female_Smoker[Younger_Adult_Female_Smoker["smoker"]==1]
YFS = Younger_Adult_Female_Smoker.groupby("region")["charges"].mean().sort_values(ascending=False).round()
YFS
```
